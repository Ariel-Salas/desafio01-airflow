```python
#+20%
#Cambiar a datetime para despu√©s insertar en la BD correspondiente
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
# df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')



# def parse_date_safe(date_str):
#     try:
#         return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce')
#     except Exception:
#         return pd.NaT

# # Aplicar a ambas columnas de forma segura
# df['Order Date'] = df['Order Date'].apply(parse_date_safe)
# df['Ship Date'] = df['Ship Date'].apply(parse_date_safe)

# print(df[['Order Date', 'Ship Date']].dtypes)  # Ambas deben ser datetime64[ns]
# print("\nFechas no v√°lidas (NaT):")
# print(df[df['Order Date'].isna() | df['Ship Date'].isna()])



import pandas as pd
import string

def clean_date_columns(df, date_columns, date_format='%d-%m-%Y', export_invalid=False):
    """
    Limpia columnas de fechas en un DataFrame.

    Par√°metros:
    - df: DataFrame de pandas
    - date_columns: lista de columnas a limpiar ['Order Date', 'Ship Date']
    - date_format: formato esperado de la fecha (por defecto: '%d-%m-%Y')
    - export_invalid: si es True, exporta las fechas inv√°lidas a archivos CSV

    Retorna:
    - df modificado con columnas parseadas y flags de caracteres raros
    - resumen con conteo de fechas inv√°lidas y caracteres raros por columna
    """
    def parse_date_safe(date_str):
        try:
            return pd.to_datetime(date_str, format=date_format, errors='coerce')
        except Exception:
            return pd.NaT

    def detect_weird_chars(s):
        if pd.isna(s):
            return False
        allowed = set(string.digits + '-')  # Solo d√≠gitos y guiones
        return any(c not in allowed for c in str(s).strip())

    summary = {}

    for col in date_columns:
        print(f"\nüîç Procesando columna: {col}")
        original_dates = df[col].copy()
        df[col] = df[col].apply(parse_date_safe)

        # Fechas no v√°lidas
        invalid_mask = df[col].isna()
        invalid_values = original_dates[invalid_mask].unique()
        summary[col] = {
            'invalid_dates_count': invalid_mask.sum(),
            'weird_chars_count': 0
        }

        if len(invalid_values) > 0:
            print(f"üö® Fechas no v√°lidas encontradas en '{col}':")
            print(invalid_values)
            if export_invalid:
                pd.Series(invalid_values).to_csv(f'invalid_dates_{col}.csv', index=False)

        # Caracteres raros
        df[f'{col} Weird'] = original_dates.apply(detect_weird_chars)
        weird_values = original_dates[df[f'{col} Weird']]
        if not weird_values.empty:
            print(f"üìõ Fechas con caracteres raros en '{col}':")
            print(weird_values.unique())
            summary[col]['weird_chars_count'] = df[f'{col} Weird'].sum()
            if export_invalid:
                weird_values.to_csv(f'weird_chars_{col}.csv', index=False)

    print("\nüìÖ Tipos de datos despu√©s de limpieza:")
    print(df[date_columns].dtypes)
    
    return df, summary


# üëá Llamada directa a la funci√≥n para que se ejecute todo y se impriman resultados
df, resumen = clean_date_columns(df, ['Order Date', 'Ship Date'])
print(resumen)
```

    discount DECIMAL(5, 2),

    rating DECIMAL(3, 1),
    FOREIGN KEY (order_id) REFERENCES Orders(order_id),

);



Clientes (Customers) a Pedidos (Orders): Relaci√≥n de uno a muchos.

Un cliente puede realizar varios pedidos, pero cada pedido pertenece solo a un cliente.

FK: Orders.customer_id es clave for√°nea que referencia a Customers.customer_id.

Pedidos (Orders) a Detalles del Pedido (Order_Details): Relaci√≥n de uno a muchos.

Un pedido puede tener varios productos (detalles de productos en el pedido), pero cada detalle de pedido pertenece a un solo pedido.

FK: Order_Details.order_id es clave for√°nea que referencia a Orders.order_id.

Productos (Products) a Detalles del Pedido (Order_Details): Relaci√≥n de uno a muchos.

Un producto puede estar presente en m√∫ltiples pedidos (detalles de pedidos), pero cada detalle de pedido pertenece a un solo producto.

FK: Order_Details.product_id es clave for√°nea que referencia a Products.product_id.
