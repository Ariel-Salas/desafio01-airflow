```python

#Sales
import pandas as pd
import re
from typing import Union

# Expresión regular precompilada para eliminar símbolos no numéricos (excepto ., - y ,)
_non_numeric_re = re.compile(r'[^\d.,-]+')

def clean_sales_to_float(value: Union[str, int, float, None]) -> float:
    """
    Limpia y convierte un valor a float para columnas de tipo 'ventas'.

    Soporta:
    - Símbolos monetarios (€, $, etc.)
    - Separadores europeos ("1.234,56") y americanos ("1,234.56")
    - Valores negativos y con espacios
    - Preserva NaN para su tratamiento posterior

    Args:
        value: El valor individual de la celda (puede ser str, int, float o None)

    Returns:
        float: Valor limpio o NaN si no se puede convertir
    """
    if pd.isna(value):
        return value  # Preserva NaN

    if isinstance(value, (int, float)):
        return float(value)

    if isinstance(value, str):
        cleaned = _non_numeric_re.sub('', value.strip())

        if ',' in cleaned and '.' in cleaned:
            cleaned = cleaned.replace('.', '').replace(',', '.')
        elif ',' in cleaned:
            parts = cleaned.split(',')
            if len(parts) == 2 and len(parts[1]) <= 2:
                cleaned = cleaned.replace(',', '.')
            else:
                cleaned = cleaned.replace(',', '')

        try:
            return float(cleaned)
        except ValueError:
            return float('nan')
    
    return float('nan')


df['Sales'] = df['Sales'].apply(clean_sales_to_float)

# Verificación rápida post-limpieza
print("\n🔍 Resumen post-limpieza:")
print(f"Tipo de datos: {df['Sales'].dtype}")  # Debería ser float64
print(f"Valores nulos: {df['Sales'].isna().sum()}")  # NaN preservados
print(f"Primeros ejemplos:\n{df['Sales'].head()}")

# Verifica si hay valores negativos 
negativos = df[df['Sales'] < 0]
if not negativos.empty:
    print("\n⚠️ Valores negativos detectados en 'Sales':")
    print(negativos)
else:
    print("\n✅ No se encontraron valores negativos en 'Sales'")
```

```python
#+20%
#Cambiar a datetime para después insertar en la BD correspondiente
# df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
# df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')



# def parse_date_safe(date_str):
#     try:
#         return pd.to_datetime(date_str, format='%d-%m-%Y', errors='coerce')
#     except Exception:
#         return pd.NaT

# # Aplicar a ambas columnas de forma segura
# df['Order Date'] = df['Order Date'].apply(parse_date_safe)
# df['Ship Date'] = df['Ship Date'].apply(parse_date_safe)

# print(df[['Order Date', 'Ship Date']].dtypes)  # Ambas deben ser datetime64[ns]
# print("\nFechas no válidas (NaT):")
# print(df[df['Order Date'].isna() | df['Ship Date'].isna()])



import pandas as pd
import string

def clean_date_columns(df, date_columns, date_format='%d-%m-%Y', export_invalid=False):
    """
    Limpia columnas de fechas en un DataFrame.

    Parámetros:
    - df: DataFrame de pandas
    - date_columns: lista de columnas a limpiar ['Order Date', 'Ship Date']
    - date_format: formato esperado de la fecha (por defecto: '%d-%m-%Y')
    - export_invalid: si es True, exporta las fechas inválidas a archivos CSV

    Retorna:
    - df modificado con columnas parseadas y flags de caracteres raros
    - resumen con conteo de fechas inválidas y caracteres raros por columna
    """
    def parse_date_safe(date_str):
        try:
            return pd.to_datetime(date_str, format=date_format, errors='coerce')
        except Exception:
            return pd.NaT

    def detect_weird_chars(s):
        if pd.isna(s):
            return False
        allowed = set(string.digits + '-')  # Solo dígitos y guiones
        return any(c not in allowed for c in str(s).strip())

    summary = {}

    for col in date_columns:
        print(f"\n🔍 Procesando columna: {col}")
        original_dates = df[col].copy()
        df[col] = df[col].apply(parse_date_safe)

        # Fechas no válidas
        invalid_mask = df[col].isna()
        invalid_values = original_dates[invalid_mask].unique()
        summary[col] = {
            'invalid_dates_count': invalid_mask.sum(),
            'weird_chars_count': 0
        }

        if len(invalid_values) > 0:
            print(f"🚨 Fechas no válidas encontradas en '{col}':")
            print(invalid_values)
            if export_invalid:
                pd.Series(invalid_values).to_csv(f'invalid_dates_{col}.csv', index=False)

        # Caracteres raros
        df[f'{col} Weird'] = original_dates.apply(detect_weird_chars)
        weird_values = original_dates[df[f'{col} Weird']]
        weird_count = df[f'{col} Weird'].sum()
        summary[col]['weird_chars_count'] = weird_count

        if weird_count > 0:
            print(f"📛 Fechas con caracteres raros en '{col}':")
            print(weird_values.unique())
            if export_invalid:
                weird_values.to_csv(f'weird_chars_{col}.csv', index=False)
        else:
            print(f"✅ Sin caracteres raros en '{col}'.")


    print("\n📅 Tipos de datos después de limpieza:")
    print(df[date_columns].dtypes)
    
    return df, summary


# Llamada directa a la función para que se ejecute todo y se impriman resultados
df, resumen = clean_date_columns(df, ['Order Date', 'Ship Date'])
print(resumen)
```

CREATE TABLE Customers (
    customer_id VARCHAR(50) PRIMARY KEY,

    segment VARCHAR(50),

CREATE TABLE Orders (
    order_id VARCHAR(50) PRIMARY KEY,
    order_date DATE,
    ship_date DATE,
    ship_mode VARCHAR(100),
    customer_id VARCHAR(50),
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id)
);

CREATE TABLE Products (
    product_id VARCHAR(50) PRIMARY KEY,
    category VARCHAR(100),
    sub_category VARCHAR(100),
    product_name VARCHAR(255)
);

CREATE TABLE Order_Details (
    order_detail_id INT AUTO_INCREMENT PRIMARY KEY,
    order_id VARCHAR(50),
    product_id VARCHAR(50),
    sales DECIMAL(10, 2),
    quantity INT,
    discount DECIMAL(5, 2),
    profit DECIMAL(10, 2),
    rating DECIMAL(3, 1),
    FOREIGN KEY (order_id) REFERENCES Orders(order_id),
    FOREIGN KEY (product_id) REFERENCES Products(product_id)
);



Clientes (Customers) a Pedidos (Orders): Relación de uno a muchos.

Un cliente puede realizar varios pedidos, pero cada pedido pertenece solo a un cliente.

FK: Orders.customer_id es clave foránea que referencia a Customers.customer_id.

Pedidos (Orders) a Detalles del Pedido (Order_Details): Relación de uno a muchos.

Un pedido puede tener varios productos (detalles de productos en el pedido), pero cada detalle de pedido pertenece a un solo pedido.

FK: Order_Details.order_id es clave foránea que referencia a Orders.order_id.

Productos (Products) a Detalles del Pedido (Order_Details): Relación de uno a muchos.

Un producto puede estar presente en múltiples pedidos (detalles de pedidos), pero cada detalle de pedido pertenece a un solo producto.

FK: Order_Details.product_id es clave foránea que referencia a Products.product_id.




  Customers "1" -- "0..*" Orders : "realiza"
    Orders "1" -- "0..*" Order_Details : "contiene"
    Products "1" -- "0..*" Order_Details : "aparece en"
